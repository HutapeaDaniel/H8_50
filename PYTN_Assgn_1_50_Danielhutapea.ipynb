```python
import numpy as np
import pandas as pd
```

```python
data_london_df = pd.read_csv('./dataset/london_crime_by_lsoa.csv')
print ('Data read into a pandas dataframe!')
```

```python
data_total = data_london_df.groupby(['year','month'],axis=0)['value'].sum()
data_total
```

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use(['ggplot'])
```

```python
line_chart = data_total
line_chart.plot(kind='line')
plt.title('London Crime Cases')
plt.ylabel('Number of Cases')
```

case: pie chart break down dari city total crime paling besar siapa?
    categori dengan total crimenya paling besar itu apa?

```python
df_lc = data_london_df
#df_lc.set_index('borough',inplace=True)
#years = list (range(2008,2016))
df_lc_total = df_lc.groupby('borough',axis=0)['value'].sum()
df_lc_total 
```

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use(['ggplot'])
```

```python
df_lc_total.plot(kind='pie',figsize=(14,8),autopct='%1.1f%%',labels=None)
plt.title('London Crime')
plt.axis('equal')
plt.legend(labels=df_lc_total.index, loc='upper left')
plt.show()
```
